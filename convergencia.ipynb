```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convergencia(historical, best):

    mean = np.mean(historical,axis=0)

    fig, ax = plt.subplots(figsize=(10,3),dpi=150)

    ax.plot(mean, color='orange', label='average')
    ax.plot(best, color='#1f77b4', label='best')
    ax.set_title('Gráfico convergencia')
    ax.set_xlabel('Iteración')
    ax.grid()
    ax.legend()

    plt.show()

    return
```

```python
convergencia([np.random.rand(50),np.random.rand(50)],np.random.rand(50))
```
